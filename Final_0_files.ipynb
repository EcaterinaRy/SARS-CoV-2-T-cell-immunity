This notebook results in files that contain immunogenic epitopes with VOCs and post-VOCs mutations, their affinity to HLA and frequency of HLA in different countries

files needed (located in the same directory as the code if not stated otherwise):
 - aa_seq_wuhan.csv, omicron_sequences.csv,  delta_sequences.csv (those are the files with reference sequencies of Wuhan, Delta and Omicron strains.
  - tcell_table_export_1701167238.csv, tcell_table_export_1701167293.csv, tcell_table_export_1701167144.csv (those are the files with Wuhan, Delta and omicron immunogenic epitopes from IEDB)
  - ReferenceWuhan.one_line.fasta
  - world_hla_alleles.csv (with world HLA alleles frequencies)
  - HLAs.csv (with countries HLA allele frequencies)
  - netMHCpan output (yours or netMHCpan/all_netMHCpan.csv file)

folders needed:
  - country, state, city (files in those directories differ by the HLA frequency datasets used)

```python
def fix_header_MHCpan(file_name):
    snp_MHC_names =pd.read_csv(file_name, sep='\t', header = [1], nrows=3)
    snp_MHC_hlas =pd.read_csv(file_name, sep='\t', header = 0, nrows=2).columns
    hlas = ['']
    hlas+=[i for i in snp_MHC_hlas if i.startswith('HLA') for j in range(2)]
    to_del = [i for i in snp_MHC_names.columns if ('Pos'in i) | ('ID' in i) | ('NB' in i)| ('core' in i)| ('icore' in i)|('Ave' in i)]
    names_col = [i for i in snp_MHC_names.columns if i not in to_del]
    my_list = []
    for chunk in pd.read_csv(file_name, sep='\t', header = 1, usecols = names_col, chunksize=50000):
        chunk[chunk.select_dtypes(np.float64).columns] = chunk.select_dtypes(np.float64).astype(np.float32)
        my_list.append(chunk)
    snp_MHC = pd.concat(my_list, axis= 0)
    del my_list
    new = [m+' ' + str(n) for m,n in zip(hlas,snp_MHC.columns)]
    new_2 = [re.sub('\.\d*', '', i) for i in new]
    snp_MHC.columns = new_2
    del snp_MHC_names, snp_MHC_hlas, new 
    return snp_MHC
```

```python
merged_new = []
def fromNetMHCpan(table):
    tables = []
    for hla in set(table['Allele Name'].values):        
        table_before_hla = table[table['Allele Name'] == hla]
        columns_netMHC = [' Peptide']
        columns_netMHC += netMHC.columns[netMHC.columns.str.startswith(hla)].to_list()
        table_netMHC = netMHC[columns_netMHC].drop_duplicates()
        table_before_hla_wt = table_before_hla.merge(table_netMHC, left_on = 'peptide', right_on = ' Peptide', how = 'left')
        table_before_hla_wt.rename(columns={hla + ' EL_Rank':'wt_EL_Rank', hla + ' BA_Rank':'wt_BA_Rank'}, inplace=True)

        table_before_hla_wt_mut = table_before_hla_wt.merge(table_netMHC, left_on = 'peptide_mut', right_on = ' Peptide', how = 'left')
        table_before_hla_wt_mut.rename(columns={hla + ' EL_Rank':'mut_EL_Rank', hla + ' BA_Rank':'mut_BA_Rank'}, inplace=True)
        tables.append(table_before_hla_wt_mut)
        #print(table_before)
    all_country_after = pd.concat(tables)
    #print(all_country_after.shape, table.shape)
    all_country_after = all_country_after.drop([' Peptide_x', ' Peptide_y'], axis=1)
    return all_country_after    

#make binding categories
def make_mark_EL(row):
    if (row['mut_EL_Rank'] <=2) & (row['wt_EL_Rank'] <=2) & (row['mut_EL_Rank'] >= 1.5*row['wt_EL_Rank']):
        return 'Decreased binding'
    elif (row['wt_EL_Rank'] <=2) & (row['mut_EL_Rank'] < row['wt_EL_Rank'])& (1.5*row['mut_EL_Rank'] <= row['wt_EL_Rank']):
        return 'Increased binding'
    elif (row['mut_EL_Rank'] >2) & (row['wt_EL_Rank'] <=2):
        return 'Lost binding'
    elif row['wt_EL_Rank'] >2:
        return 'No binding'
    else:
        return 'Same'
def make_mark_EL_wt(row):
    if (row['wt_EL_Rank'] >2):
        return 'No binding'
    elif (row['wt_EL_Rank'] <=2) & (row['wt_EL_Rank'] >0.5):
        return 'weak binder'
    elif (row['wt_EL_Rank'] <=0.5):
        return 'strong binder'
```
